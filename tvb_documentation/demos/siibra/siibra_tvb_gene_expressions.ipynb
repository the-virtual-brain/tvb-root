### Gene expressions

#### • Gene expressions of a particular **gene** can be searched in each region
#### • Inside each region multiple *gene expressions* were measured (for a *single gene*), corresponding with *multiple donors*
#### • For each donor, multiple probes (measurments) were taken
#### • Each probe contains a list of gene expressions corresponding to multiple samples being taken from the same probe
#### **gene -> region -> donor -> probe -> sample**

#### => A ConnectivityMeasure could be 

```python
root_reg = jubrain.regiontree
```

```python
# get all regions
all_reg = [region for region in jubrain.regiontree]
len(all_reg)
```

```python
# get list of all genes
gene_names = list(siibra.features.gene_names)
```

```python
# find the gene expression for the first gene in the list in all regions
reg_features = []
for reg in all_reg[:10]:
    features = siibra.get_features(
    reg, siibra.modalities.GeneExpression,
    gene=gene_names[0],
    maptype=siibra.commons.MapType.CONTINUOUS,
    threshold_continuous=0.2)
    reg_features.append(features)
```
